```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define o valor de x e o valor exato de e^x
x = 1.5
ex = np.exp(x)
```

```python
# define o número de termos da série de Taylor e o valor de epsilon
N = 5
eppara = 0.5 * 10**(2-N)
```

```python
# inicializa vetores
Est = np.zeros(N)     # estimativas
Ept = np.zeros(N)     # erro percentual verdadeiro 
Epest = np.zeros(N)   # erro percentual estimado 
```

```python
# primeira estimativa
Est[0] = 1
Ept[0] = abs((ex - Est[0]) / ex) * 100
Epest[0] = 100
```

```python
# iterações para as demais estimativas
for i in range(1, N):
    # atualiza a estimativa
    Est[i] = Est[i-1] + (x**(i))/np.math.factorial(i)
    
    # calcula o erro percentual verdadeiro
    Ept[i] = abs((ex - Est[i]) / ex) * 100
    
    # calcula o erro percentual estimado
    Epest[i] = abs((Est[i] - Est[i-1]) / Est[i]) * 100
```

```python
# exibe a tabela de resultados
RES = np.vstack((np.arange(1, N+1), Est, Ept, Epest))
print(RES)
```

```python
# plota os gráficos dos erros
plt.plot(np.arange(1, N+1), Ept, 'b', linewidth=2)
plt.plot(np.arange(1, N+1), Epest, '-.r', linewidth=2)
plt.xlabel('Termos/Iterações')
plt.ylabel('Erros')
plt.legend(['E_pver', 'E_pest'])
plt.title('Análise de Erros')
plt.show()
```
